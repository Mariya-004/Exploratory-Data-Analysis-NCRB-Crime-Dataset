EDA-SUMMARY OF DATASET

```r
library(dplyr)
library(stringr)
library(skimr)

crime_data_list <- list(
  "2017" = crime_2017,
  "2018" = crime_2018,
  "2019" = crime_2019,
  "2020" = crime_2020,
  "2021" = crime_2021
)

inspect_dataset <- function(df, year) {


  parent_rows <- sapply(df$s_no, function(id)
    !str_detect(id, "\\.") &&
    any(str_starts(df$s_no, paste0(id, ".")))
  )

  df_clean <- df %>%
    filter(!parent_rows) %>%                              # remove hierarchical totals
    filter(!grepl("Total", category,  ignore.case = TRUE), # optional extra safeguard
           !grepl("Total", crime_head, ignore.case = TRUE))

  cat("\n==== Year:", year, "====\n")

  str(df_clean)
  print(summary(df_clean))
  skim(df_clean)

  cat("Missing values per column:\n")
  print(colSums(is.na(df_clean)))

  return(df_clean)
}

# Apply to all years
cleaned_crime_data <- lapply(names(crime_data_list), function(year) {
  inspect_dataset(crime_data_list[[year]], year)
})
names(cleaned_crime_data) <- names(crime_data_list)
```

yearwise trends(combined)

```r
datasets <- list("2017" = crime_2017,
                 "2018" = crime_2018,
                 "2019" = crime_2019,
                 "2020" = crime_2020,
                 "2021" = crime_2021)

year_summary <- data.frame(
  Year = names(datasets),
  Total_Reported = sapply(datasets, function(df) {
    df %>%
      filter(!grepl("total", tolower(crime_head))) %>%   # remove total/subtotal rows
      summarise(total = sum(cases_reported, na.rm = TRUE)) %>%
      pull(total)
  }),
  Avg_ChargeRate = sapply(datasets, function(df) mean(df$chargesheeting_rate, na.rm = TRUE)),
  Avg_Pendency   = sapply(datasets, function(df) mean(df$pendency_rate, na.rm = TRUE))
)

print(year_summary)


# Total Reported Cases
total_reported=ggplot(year_summary, aes(x = Year, y = Total_Reported, group = 1)) +
  geom_line(color = "red", size = 1.2) +
  geom_point(size = 3, color = "blue") +
  labs(title = "Yearly Total Reported Cases (2017–2021)",
       x = "Year", y = "Cases Reported")
 total_reported


# Average Charge-Sheeting Rate
average_chargesheeting=ggplot(year_summary, aes(x = Year, y = Avg_ChargeRate, group = 1)) +
  geom_line(color = "darkgreen", size = 1.2) +
  geom_point(size = 3, color = "orange") +
  labs(title = "Yearly Average Charge-Sheeting Rate (2017–2021)",
       x = "Year", y = "Charge-Sheeting Rate (%)")
average_chargesheeting
# Average Pendency Rate
average_pendency=ggplot(year_summary, aes(x = Year, y = Avg_Pendency, group = 1)) +
  geom_line(color = "blue", size = 1.2) +
  geom_point(size = 3, color = "red") +
  labs(title = "Yearly Average Pendency Rate (2017–2021)",
       x = "Year", y = "Pendency Rate (%)")
average_pendency
```

```r
# Load libraries
library(ggplot2)
library(reshape2)
library(dplyr) # Ensure dplyr is loaded for %>%
library(janitor) # Load janitor for clean_names()


# Read all datasets, clean names, standardize columns, select relevant columns, and add "Year" column
crime_2017_processed <- read.csv("2017_crime_data.csv") %>%
  clean_names() %>%
  standardize_columns() %>% # Apply the corrected standardize_columns function
  select(cases_reported, total_disposed, cases_pending_end_year) %>% # Select ONLY the columns for the heatmap
  mutate(Year = 2017)

crime_2018_processed <- read.csv("2018_crime_data.csv") %>%
  clean_names() %>%
  standardize_columns() %>% # Apply the corrected standardize_columns function
  select(cases_reported, total_disposed, cases_pending_end_year) %>% # Select ONLY the columns for the heatmap
  mutate(Year = 2018)

crime_2019_processed <- read.csv("2019_crime_data.csv") %>%
  clean_names() %>%
  standardize_columns() %>% # Apply the corrected standardize_columns function
  select(cases_reported, total_disposed, cases_pending_end_year) %>% # Select ONLY the columns for the heatmap
  mutate(Year = 2019)

crime_2020_processed <- read.csv("2020_crime_data.csv") %>%
  clean_names() %>%
  standardize_columns() %>% # Apply the corrected standardize_columns function
  select(cases_reported, total_disposed, cases_pending_end_year) %>% # Select ONLY the columns for the heatmap
  mutate(Year = 2020)

crime_2021_processed <- read.csv("2021_crime_data.csv") %>%
  clean_names() %>%
  standardize_columns() %>% # Apply the corrected standardize_columns function
  select(cases_reported, total_disposed, cases_pending_end_year) %>% # Select ONLY the columns for the heatmap
  mutate(Year = 2021)


# Combine all years
crime_all <- rbind(crime_2017_processed, crime_2018_processed, crime_2019_processed, crime_2020_processed, crime_2021_processed)

# The relevant columns are already selected and renamed in the processing steps
crime_subset <- crime_all


# Aggregate by Year (sum of all crimes in that year)
crime_yearly <- aggregate(. ~ Year, data = crime_subset, sum)

# Reshape for heatmap
crime_long <- melt(crime_yearly, id.vars = "Year")

# Plot heatmap
ggplot(crime_long, aes(x = variable, y = factor(Year), fill = value)) +
  geom_tile(color = "white") +
  scale_fill_gradient(low = "lightblue", high = "red") +
  labs(title = "Yearly Heatmap of Reported vs Disposed vs Pending Cases",
       x = "Case Type",
       y = "Year") +
  theme_minimal(base_size = 12)
```
